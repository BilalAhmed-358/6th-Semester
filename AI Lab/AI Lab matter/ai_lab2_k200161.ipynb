# Matplotlib

**Task 1:** &emsp; Create a line chart consisting of 2 separate lines each having a label of its own and one of them styled in a doted manner. Also add labels to the axes.

```python
import matplotlib.pyplot as plt

line1 = np.array([3, 8, 1, 10])
line2 = np.array([1, 3, 20, 9])
plt.plot(line1, 'o:r', label = 'Line 1', linestyle ='solid')
plt.plot(line2, 'o:b', label = 'Line 2')
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

**Task 2:** &emsp; Create a Pie Chart similar to the one given below. You can use dataset and colors of your own, but make sure that the structure is followed as is.

```python
weights = np.array([1034, 1724, 2069, 5172])
gaps = np.array([0.1, 0.1, 0.1, 0.1])
labels = ['Repairs', 'Food', 'Utilities', 'Mortgage']
plt.title('Household Expenses')
plt.pie(weights, labels = labels, startangle = 90, autopct='%1.2f%%', explode=gaps, counterclock=False)
plt.show()
```

# NLTK

```python
import nltk
nltk.download('punkt')
nltk.download("stopwords")

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize

stop_words = set(stopwords.words('english'))
text = 'Joe waited for the train. The train was late. Mary and samantha took the bus. I looked for Mary and Samantha at the bus station.'
```

**Task 1:** &emsp; Write a Python NLTK program to split the text sentence paragraph into a list of words.

```python
print('Original string:')
print(text)
print()

sentences = sent_tokenize(text)
print('Sentence-tokenized copy in a list:')
print(sentences)
print()

print('Read the list')
for sent in sentences:
  print(sent)
```

**Task 2:** &emsp; Write a Python NLTK program to create a list of words from a given string.

```python
print('Original string:')
print(text)
print()

words = word_tokenize(text)
print('List of words:')
print(words)
print()
```

**Task 3:** &emsp; Write a Python NLTK program to tokenize words, sentence wise.

```python
print('Original string:')
print(text)
print()

sentences = sent_tokenize(text)

print('Read the list')
for sent in sentences:
  tokens_per_sent = word_tokenize(sent)
  print(tokens_per_sent)
```

# Spacy

**Task 1:** &emsp; Create a syntactic dependency visualizer for a given sentence.

```python
import spacy
from spacy import displacy

sentence = "Apple is looking at buying U.K. startup for $1 billion"
nlp = spacy.load("en_core_web_sm")
doc = nlp(sentence)

displacy.render(doc, style='dep', jupyter=True, options={'distance': 90})
```

**Task 2:** &emsp; Break a given sentence into tokens each representing a single word.

```python
import spacy

nlp = spacy.blank("en")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for token in doc:
  print(token)
```
