## **Task 1:** &emsp;  Dice Simulation

Two unbiased dice are thrown once and the total score is observed. Use a simulation to find
the estimated probability that the total score is even or greater than 7?

```python
import random

def roll_dice():
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    return die1 + die2

n = 1000000
count = 0

for i in range(n):
    score = roll_dice()
    if score % 2 == 0 or score > 7:
        count += 1

prob = count / n
print("Estimated probability of score being even or greater than 7:", prob)
```

## **Task 2:** &emsp;  Balls with Replacement

A box contains 10 white balls, 20 reds and 30 greens. Draw 5 balls with replacemen. what is the probability that:
1. 3 white or 2 red
2. All 5 are the same color

```python
from itertools import product
colors = ["white", "red", "green"]
counts = [10, 20, 30]

# to calculate probability of drawing n balls given color
def probability(color, n):
    index = colors.index(color)
    ball_count = counts[index]
    total_balls = sum(counts) 
    return ball_count / total_balls ** n

# to calculate probability of given combinations
def combination_probability(combination):
    result = 1
    for color, count in combination.items():
        result *= probability(color, count)
    return result

# to generate all possible combinations of n balls
def generate_combinations(n):
    combinations = []
    # all possible counts of white balls
    for num_white in range(n + 1):
        # all possible counts of red balls
        for num_red in range(n - num_white + 1):
            # count of green balls
            num_green = n - num_white - num_red
            combination = {"white": num_white, "red": num_red, "green": num_green}
            combinations.append(combination)
    return combinations

prob_3_white_or_2_red = 0
for combination in generate_combinations(5):
    if combination.get("white", 0) == 3 or combination.get("red", 0) == 2:
        prob_3_white_or_2_red += combination_probability(combination)

print("Probability of drawing 3 white balls or 2 red balls:", prob_3_white_or_2_red)

prob_all_same_color = 0
for color in colors:
    prob_all_same_color += combination_probability({color: 5})

print("Probability of drawing all 5 balls of the same color:", prob_all_same_color)
```

## **Task 3:** &emsp;  Four Siblings Problem

The sample space for four sibling children. Each row in the sample space contains 1 of 16
possible outcomes. Every outcome represents a unique combination of four children. The
sex of each child is indicated by a letter: B for boy and G for girl. Outcomes with two boys
are marked by an arrow. Calculate the probability of two boys.

```python
sample_space = []
for child1 in ["B", "G"]:
    for child2 in ["B", "G"]:
        for child3 in ["B", "G"]:
            for child4 in ["B", "G"]:
                sample_space.append((child1, child2, child3, child4))

num_two_boys = 0
for outcome in sample_space:
    if outcome.count("B") == 2:
        num_two_boys += 1

prob_two_boys = num_two_boys / len(sample_space)
print("The probability of two boys is:", prob_two_boys)
```

## **Task 4:** &emsp;  Coin Toss Problem

A coin is tossed twice. What is the probability that at least 1 head occurs? The sample space for this experiment is S = {HH, HT, TH, TT}.

```python
sample_space = ["HH", "HT", "TH", "TT"]

num_no_heads = 0
for outcome in sample_space:
    if outcome == "TT":
        num_no_heads += 1

prob = 1 - num_no_heads / len(sample_space)
print("The probability of getting at least one head is:", prob)
```

A die is loaded in such a way that an even number is twice as likely to occur as an odd

## **Task 7:** &emsp;  Manufacturing Firm

A manufacturing firm employs three analytical plans for the design and development of a
particular product. For cost reasons, all three are used at varying times. In fact, plans 1, 2,
and 3 are used for 30%, 20%, and 50% of the products, respectively. The defect rate is
different for the three procedures as follows: P(D|P1)=0.01, P(D|P2)=0.03, P(D|P3)=0.02,
where P(D|Pj ) is the probability of a defective product, given plan j. If a random product was
observed and found to be defective, which plan was most likely used and thus responsible?

```python
P1 = 0.3
P2 = 0.2
P3 = 0.5

# using bayes' rule
P_D_given_P1 = 0.01
P_D_given_P2 = 0.03
P_D_given_P3 = 0.02

P_D = P_D_given_P1*P1 + P_D_given_P2*P2 + P_D_given_P3*P3

P_P1_given_D = P_D_given_P1 * P1 / P_D
P_P2_given_D = P_D_given_P2 * P2 / P_D
P_P3_given_D = P_D_given_P3 * P3 / P_D

print("Probability of each plan being used given a defective product:")
print("Plan 1:", P_P1_given_D)
print("Plan 2:", P_P2_given_D)
print("Plan 3:", P_P3_given_D)

# determine which plan is most likely responsible
if P_P1_given_D > P_P2_given_D and P_P1_given_D > P_P3_given_D:
    print("Plan 1 is most likely responsible")
elif P_P2_given_D > P_P1_given_D and P_P2_given_D > P_P3_given_D:
    print("Plan 2 is most likely responsible")
else:
    print("Plan 3 is most likely responsible")
```
